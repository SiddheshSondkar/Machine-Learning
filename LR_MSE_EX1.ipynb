```python
X=[1,2,3,4,5]
Y=[3,4,2,4,5]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,Y)
plt.show
```

```python
X=[1,2,3,4,5]
actual_Y=[3,4,2,4,5]
pred_Y=[2.8,3.2,3.6,4,4.4]
```

```python
plt.scatter(X,actual_Y,marker='*')
plt.plot(X,pred_Y,color='g')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(actual_Y,pred_Y)
```
